```python

'''

# 🔎 1️⃣ Your Corpus Is Clean

Your essay:

* Has clear paragraph boundaries
* Has logical topic transitions
* Has structured flow

Recursive splitter works extremely well when:

* Text is already well-written
* Paragraphs align with semantic shifts

So RCE isn’t “smart.”
Your document is just behaving nicely.

LLM chunking doesn’t gain much because there’s nothing messy to fix.

---

# 🔎 2️⃣ LLM Chunking Often Creates Larger Chunks

LLM chunkers tend to:

* Merge related ideas
* Create broader conceptual sections

That sounds intelligent.

But for retrieval:

Broader chunk → embedding averages more concepts → semantic dilution.

Example:

Chunk contains:

* Attention mechanisms
* Transformers
* Scaling laws

Now query = “Why did attention help RNNs?”

Embedding similarity becomes weaker because vector represents multiple themes.

RCE might isolate attention discussion more tightly.

Smaller chunk = higher embedding purity.

That boosts Precision@K.

---

# 🔎 3️⃣ LLM Chunking Is Not Optimizing for Retrieval

This is the key mistake people make.

LLM chunking optimizes for:

* Human-readable sections
* Logical narrative structure

Retrieval optimizes for:

* Vector similarity alignment
* Topical isolation
* Embedding density

These are not identical goals.

Clean narrative ≠ optimal vector retrieval.

---

# 🔎 4️⃣ Over-Semantic Chunking Can Hurt

If LLM:

* Groups context heavily
* Avoids splitting subtopics

You get:

Fewer chunks
Each chunk covers broader scope

That reduces ranking sharpness.

Precision@K drops because:

Top 3 results contain more semi-relevant chunks instead of one laser-focused chunk.

---

# 🔎 5️⃣ Embedding Model Interaction

Important.

Your embedding model is likely:

* Sentence-level optimized
* Context-window limited

If LLM chunks are too large:

* Embedding becomes coarse
* Subtopic signals weaken

Recursive chunks might better match embedding model’s sweet spot.

This is model-chunk alignment.

---

# 🔎 6️⃣ LLM Judging Bias

If you’re using LLM-as-judge:

It may prefer:

* Concise chunks
* Direct topical answers

RCE might accidentally produce more direct chunks.

LLM chunking may include surrounding context that reduces perceived precision.

---

# 🔥 The Big Insight

Smarter segmentation ≠ better retrieval.

Retrieval cares about:

* Signal density
* Topical isolation
* Vector discriminability

LLM chunking often improves:

* Human interpretability
* Logical structure

But not necessarily embedding geometry.

---

# 🧠 What This Means For You

If:

RCE Precision@3 = higher
RCE nDCG@3 = higher

Then for this dataset:

RCE is the better engineering choice.

And that’s perfectly fine.

You don’t get bonus points for complexity.

---

# ⚖️ When LLM Chunking Would Win

* Messy transcripts
* Legal documents
* No paragraph boundaries
* Topic blending inside paragraphs

Not clean essays like yours.
```

```python
CORRECTNESS_JUDGE_PROMPT = """
You are a strict evaluation system.

You are given:
1. The original user query.
2. The gold (reference) answer.
3. The model-generated answer.
4. Required facts that must appear in the answer.
5. Disallowed facts that must NOT appear in the answer.

Your task:
Evaluate whether the model answer is correct.

Evaluation Rules:
- The answer must contain ALL required facts.
- The answer must NOT contain any disallowed facts.
- Numeric values must match exactly.
- Missing required facts = Incorrect.
- Any disallowed fact present = Incorrect.
- Paraphrasing is allowed only if meaning is equivalent.
- Do NOT be lenient.

Return ONLY valid JSON in this format:

{{
  "correct": true | false,
  "missing_required_facts": [],
  "present_disallowed_facts": [],
  "reason": "short explanation"
}}

Query:
{query}

Gold Answer:
{gold_answer}

Model Answer:
{model_answer}

Required Facts:
{required_facts}

Disallowed Facts:
{disallowed_facts}
"""
```

```python
FAITHFULNESS_JUDGE_PROMPT = """
You are a strict reasoning evaluation system.

You are given:
1. A user query.
2. Retrieved context.
3. The model-generated answer.

Your task:
Determine whether the model's reasoning and conclusions strictly follow from the retrieved context.

Rules:
- The answer must not introduce unstated assumptions.
- The answer must not exaggerate implications.
- Logical conclusions must be directly supported by the context.
- Minor paraphrasing is allowed.
- If reasoning extends beyond context, mark as false.

Return ONLY valid JSON:

{{
  "faithful": true | false,
  "reason": "short explanation"
}}
Query:
{query}

Retrieved Context:
{context}

Model Answer:
{answer}
"""


RELEVANCE_JUDGE_PROMPT = """
You are a strict relevance evaluation system.

You are given:
1. A user query.
2. A model-generated answer.

Your task:
Determine whether the answer directly addresses the user’s question.

Rules:
- The answer must focus on what is being asked.
- Extra unrelated technical details reduce relevance.
- If the answer does not directly answer the question, mark false.

Return ONLY valid JSON:
{{
  "relevant": true | false,
  "reason": "short explanation"
}}

Query:
{query}

Model Answer:
{answer}
"""
```

```python
import json

def run_full_rag_evaluation():
    final_results = []

    for item in Query:

        print(f"\nProcessing {item['id']}")

        # =========================================================
        # 1️⃣ RETRIEVAL
        # =========================================================
        retrieved_chunks = multi_query_retriever(item['query'], 3, gen_db)
        context = "\n\n".join([chunk for chunk in retrieved_chunks])

        # =========================================================
        # 2️⃣ GENERATION
        # =========================================================
        generation_prompt = GENERATION_PROMPT.format(
            query=item["query"],
            context=context
        )

        answer = llm_query.invoke(generation_prompt).content.strip()

        # =========================================================
        # 3️⃣ CLAIM EXTRACTION
        # =========================================================
        extraction_prompt = CLAIM_EXTRACTION_PROMPT.format(answer=answer)

        try:
            extraction_response = llm.invoke(extraction_prompt).content
            claims = json.loads(extraction_response).get("claims", [])
        except:
            claims = []

        # =========================================================
        # 4️⃣ CLAIM VERIFICATION (Groundedness + Hallucination)
        # =========================================================
        labels = []

        for claim in claims:
            verification_prompt = CLAIM_VERIFICATION_PROMPT.format(
                query=item["query"],
                context=context,
                claim=claim
            )

            try:
                verification_response = llm_query.invoke(verification_prompt).content
                label = json.loads(verification_response)["label"]
            except:
                label = "Unsupported"

            labels.append(label)

        if len(labels) == 0:
            groundedness_score = 0.0
            hallucination_rate = 0.0
        else:
            supported_count = sum(1 for l in labels if l == "Supported")
            unsupported_count = sum(1 for l in labels if l == "Unsupported")

            groundedness_score = supported_count / len(labels)
            hallucination_rate = unsupported_count / len(labels)

        # =========================================================
        # 5️⃣ CORRECTNESS JUDGE
        # =========================================================
        judge_prompt = CORRECTNESS_JUDGE_PROMPT.format(
            query=item["query"],
            gold_answer=item["gold_answer"],
            model_answer=answer,
            required_facts=item["required_facts"],
            disallowed_facts=item["disallowed_facts"]
        )

        try:
            judge_response = llm.invoke(judge_prompt).content
            judge_json = json.loads(judge_response)

            correctness = judge_json.get("correct", False)
            missing_required = judge_json.get("missing_required_facts", [])
            disallowed_present = judge_json.get("present_disallowed_facts", [])
            correctness_reason = judge_json.get("reason", "")
        except:
            correctness = False
            missing_required = ["Judge parsing failed"]
            disallowed_present = []
            correctness_reason = "Judge output parsing error"

        # =========================================================
        # 6️⃣ FAITHFULNESS JUDGE
        # =========================================================
        faithfulness_prompt = FAITHFULNESS_JUDGE_PROMPT.format(
            query=item["query"],
            context=context,
            answer=answer
        )

        try:
            faithfulness_response = llm_query.invoke(faithfulness_prompt).content
            faithfulness_json = json.loads(faithfulness_response)

            faithful = faithfulness_json.get("faithful", False)
            faithfulness_reason = faithfulness_json.get("reason", "")
        except:
            faithful = False
            faithfulness_reason = "Faithfulness parsing error"

        # =========================================================
        # 7️⃣ RELEVANCE JUDGE
        # =========================================================
        relevance_prompt = RELEVANCE_JUDGE_PROMPT.format(
            query=item["query"],
            answer=answer
        )

        try:
            relevance_response = llm.invoke(relevance_prompt).content
            relevance_json = json.loads(relevance_response)

            relevant = relevance_json.get("relevant", False)
            relevance_reason = relevance_json.get("reason", "")
        except:
            relevant = False
            relevance_reason = "Relevance parsing error"

        # =========================================================
        # 8️⃣ STORE EVERYTHING
        # =========================================================
        result = {
            "query_id": item["id"],
            "query": item["query"],
            "model_answer": answer,

            # Groundedness
            "claims": claims,
            "labels": labels,
            "groundedness_score": groundedness_score,
            "hallucination_rate": hallucination_rate,

            # Correctness
            "correct": correctness,
            "missing_required_facts": missing_required,
            "present_disallowed_facts": disallowed_present,
            "correctness_reason": correctness_reason,

            # Faithfulness
            "faithful": faithful,
            "faithfulness_reason": faithfulness_reason,

            # Relevance
            "relevant": relevant,
            "relevance_reason": relevance_reason
        }

        final_results.append(result)

        print(result)

    return final_results
```
